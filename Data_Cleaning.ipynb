## **5. Rebuild missing data** <a name="step5"></a>

**The review_title, review_description, and travel_date columns have some missing data; let's look at the missing rows and figure out how to deal with them.**

```python
pd.set_option('display.max_rows', None)
df_all[df_all.isnull().any(axis=1)]
```

**Because the review_title and review_description are the most important data in our dataset, we will delete all rows that do not contain this data.**

```python
df_all= df_all.dropna(subset=['review_title', 'review_description'])
```

```python
df_all.isna().sum()
```

**Now that only the travel_date column is missing data, we will fill it by review_date because it may be the same or more relevant to the travel_date.**

```python
df_all.loc[(pd.isnull(df_all.travel_date)), 'travel_date'] = 'Date of travel: '+str(df_all['review_date'].str.split(' ')[0])+' '+str(df_all['review_date'].str.split(' ')[2])
```

## **7. Cleaning text columns** <a name="step7"></a>

#### **A. The global information dataframe**

Use airline_name column as the index

```python
dfGeneral.set_index('airline_name', inplace=True)
```

Number of reviews column is a string column need to be converted to integer

```python
dfGeneral['reviews'] = dfGeneral['reviews'].str.split(' ').str[0].str.replace(',', '').astype(float)
dfGeneral
```

```python
dfGeneral.info()
```

Rename the reviews column to reviews_number

```python
dfGeneral.rename(columns={'reviews': 'reviews_number'}, inplace=True)
dfGeneral.columns
```

To easily create plots, extract the years from the text in travelers_choice_winner column

```python
# convert list to string
dfGeneral['travelers_choice_winner']= dfGeneral['travelers_choice_winner'].astype(str).str.replace('\[|\]|\'', '')
# extract the years from the string
for index,item in dfGeneral.iterrows():
    dfGeneral['travelers_choice_winner'][index]=([int(s) for s in dfGeneral['travelers_choice_winner'][index].split() if s.isdigit()])
```

```python
dfGeneral.head()
```

#### **B. The airlines dataframe**

Extract year and month from the travel_date and review_date to add new column, this will allow us to do some trend analysis later.

```python
temp_review= df_all['review_date']
df_all.insert(4, "review_month", temp_review.str.split(' ').str[0].str.strip())
df_all.insert(5, "review_year", temp_review.str.split(' ').str[2].str.strip())
```

```python
# Check unique year and month values if contains not correct values
print('Unique year values of review_year column are: ', *df_all.review_year.unique(), sep = ', ')
print('Unique month values of review_year column are: ', *df_all.review_month.unique(), sep = ', ')
```

```python
# conver month name to number
month_dict = dict((v, k) for k, v in enumerate(calendar.month_name))
df_all.review_month= df_all.review_month.map(month_dict)
```

```python
df_all = df_all.drop('review_date', axis=1)
```

```python
temp_travel= df_all['travel_date'].str.split(':').str[1].str.strip()
df_all.insert(7, "travel_month", temp_travel.str.split(' ').str[0].str.strip())
df_all.insert(8, "travel_year", temp_travel.str.split(' ').str[1].str.strip())
```

```python
# Check unique year and month values if contains not correct values
print('Unique year values of travel_year column are: ', *df_all.travel_year.unique(), sep = ', ')
print('Unique month values of travel_month column are: ', *df_all.travel_month.unique(), sep = ', ')
```

**As we can see above, there are errors in both columns that need to be fixed. Error data in the year column will be replaced with the review_year value from the same row. The month column has an error, which is ['October,' which will be modified to October.**

```python
df_all.loc[df_all['travel_month'] == "['October',", ['travel_month']]= 'October'
```

```python
incor_years= ["'6,',", '2563', '2562', '2561']
#df_all.loc[df_all['travel_year'].isin(incor_years), ['travel_year']] = df_all['review_year']
for i in incor_years:
  df_all.loc[df_all['travel_year']==i, ['travel_year']] = df_all['review_year']
```

```python
# conver month name to number
df_all.travel_month= df_all.travel_month.map(month_dict)
```

```python
# Check again unique year and month values if contains not correct values
print('Unique year values of travel_year column are: ', *df_all.travel_year.unique(), sep = ', ')
print('Unique month values of travel_month column are: ', *df_all.travel_month.unique(), sep = ', ')
```

```python
df_all = df_all.drop('travel_date', axis=1)
```
